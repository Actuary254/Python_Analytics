# **Label Encoding**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load the Excel file into a DataFrame
file_path = "/content/Test File.xlsx"
df = pd.read_excel(file_path)

# Define categorical columns
categorical_columns = ["Country", "Category", "Manufacture"]

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Create a copy of the original DataFrame
df1 = df.copy()

# Perform label encoding for each categorical column
for col in categorical_columns:
    if col in df1.columns:
        df1[col + '_encoded'] = label_encoder.fit_transform(df1[col])

# Display the encoded DataFrame
print("Encoded DataFrame:")
print(df1)
```

```python
# View data types of each column in the DataFrame
print("Data types of each column:")
print(df.dtypes)
```

```python
# Exclude specified columns
columns_to_exclude = ["Country ", "Category ", "Manufacture"]
df_subset = df.drop(columns=columns_to_exclude)

# Convert remaining columns to integers
df_subset = df_subset.astype(int)

# View data types after conversion
print("Data types of each column after conversion:")
print(df_subset.dtypes)
```

# **Data Visualization**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a heatmap for the remaining variables
plt.figure(figsize=(20, 12))
sns.heatmap(df_subset.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Remaining Variables")
plt.show()
```

```python
# Compute the correlation matrix
correlation_matrix = df_subset.corr()

# Display the correlation matrix as a table
print("Correlation Matrix:")
print(correlation_matrix)
```
